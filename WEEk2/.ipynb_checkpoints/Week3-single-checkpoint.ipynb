```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as op
```

```python
filename = 'ex2data1.txt'
data = np.loadtxt(filename, delimiter=',')#b=np.linalg.inv(a)
```

```python
#plt data
#from mpl_toolkits.mplot3d import Axes3D
#ax = plt.subplot(111, projection='3d')
ax = plt.subplot(111)
X_0 = data[np.where(data==0)[0]]
X_1 = data[np.where(data==1)[0]]
ax.scatter(X_0[:,0:1], X_0[:, 1:2], marker = 'x', c='r', label='NoAdmit')
ax.scatter(X_1[:,0:1], X_1[:, 1:2], marker = 'o',c='b', label='Admit')
plt.legend(loc=0)
plt.xlabel('test 1')
plt.ylabel('test 2')
plt.show()
```

```python
X_no1 = data[:,0:2]
y = data[:, 2:]
X = np.c_[np.ones((X_no1.shape[0], 1)), X_no1]
```

```python
theta = np.zeros((X.shape[1], 1))
```

```python
def sigmoid(theta, X):
    z = X.dot(theta)
    g = 1/(1+np.exp(-z))
    return g
```

```python
def costfunction(theta, X, y):
    m = len(y)
    h = sigmoid(theta, X)
    J = 1/m*(-y.T.dot(np.log(h))-(1-y).T.dot(np.log(1-h)))
    return J
```

```python
costfunction(theta, X, y)
```

```python
def gradientdescent(theta, X, y):
    m = len(X)
    h = sigmoid(theta, X)
    grad = 1/m*(X.T.dot(h-y))
    return grad.flatten()
    
```

```python
result = op.minimize(fun=costfunction, x0=theta, args=(X, y), method='TNC')# jac=gradent)
theta_optim = result.x
theta_optim
```

```python
ax1 = plt.subplot(111)
X_0 = data[np.where(data==0)[0]]
X_1 = data[np.where(data==1)[0]]
ax1.scatter(X_0[:,0:1], X_0[:, 1:2], marker = 'x', c='r', label='NoAdmit')
ax1.scatter(X_1[:,0:1], X_1[:, 1:2], marker = 'o',c='b', label='Admit')
min_x = X[:, 1].min(0)-2
max_x = X[:, 1].max(0)-2
plt_x = np.array([min_x, max_x])
min_y = -1/theta_optim[2]*(theta_optim[0] + theta_optim[1]*min_x)
max_y = -1/theta_optim[2]*(theta_optim[0] + theta_optim[1]*max_x)
plt_y = np.array([min_y, max_y])
ax1.plot(plt_x, plt_y)

plt.legend(loc=0)
plt.xlabel('test 1')
plt.ylabel('test 2')
plt.show()
```

```python
#prob = sigmoid(np.array([1,45,85]).dot(optimal_theta))
def predict(theta, X):
    m = X.shape[0]
    p = np.zeros((m, 1))
    k = np.where(sigmoid(theta, X) >= 0.5)[0]
    p[k]= 1
    return p
p = predict(theta_optim, X)
acc = (p==y).mean()
print(acc)
```
