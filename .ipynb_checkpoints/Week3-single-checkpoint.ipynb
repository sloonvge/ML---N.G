```python
X = np.c_[np.ones((len(data1),1)), data1[:, 0:1]]
y = data1[:, 1:]
alpha = 0.01
theta = np.zeros((2,1))
iterations = 1500
```

```python
def comcost(X, y, theta):
    m = len(y)
    J = sum((X.dot(theta)-y)**2)/(2*m)
    return J
```

```python
cost = comcost(X, y, theta)
cost
```

```python
cost = comcost(X, y, [[-1],
                          [2]])
cost
```

```python
def gradientDescent(X, y, theta, alpha, iter_num):
    m = len(y)
    J_history = []
    theta_s = theta
    i = 0
    while i < iter_num:
        theta[0] = theta[0] - sum(X.dot(theta_s)-y)*alpha/m
        theta[1] = theta[1] - sum((X.dot(theta_s)-y)*X[:,1:])*alpha/m
        
        cost = comcost(X, y, theta_s)
        theta_s = theta
        J_history.append(cost[0])
        i +=1
        
    return theta,J_history
    
```

```python
theta = gradientDescent(X, y, theta, alpha, iterations)[0]
```

```python
predict1 = np.array([[1, 3.5]]).dot(theta)
predict1*10000
```
